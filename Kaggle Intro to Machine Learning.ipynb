# Data Exploration Using Kaggle Data

## Dataset used from [Kaggle.com](https://www.kaggle.com/datasets)

```python
# import Iowa housing data

import pandas as pd

iowa_housing_file_path = './train.csv'
home_data = pd.read_csv(iowa_housing_file_path)

# describe data
home_data.describe()
```

```python
# First ML model using Melbourne housing data

melbourne_housing_file_path = './melb_data.csv'
melb_data = pd.read_csv(melbourne_housing_file_path)
melb_data.columns
```

```python
# The Melbourne dataset has some missing values. For the purpose of this learning we will drop all missing values.
# Though in reality we should investigate the missing values further before deciding to drop them.

melb_data = melb_data.dropna(axis=0)
```

```python
# we want to predict price so we will set that as our 'y'
y = melb_data.Price

# Features we are interested in to predict price
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X = melb_data[melbourne_features]

X.head()
```

```python
# Build our model using Scikit Learn

from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
melbourne_model = DecisionTreeRegressor(random_state=1)

# Fit model
melbourne_model.fit(X, y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(melbourne_model.predict(X.head()))
```

```python
# print the list of columns in the dataset to find the name of the prediction target
print(home_data.columns)

# Set target variable to SalePrice
y_iowa = home_data.SalePrice

# Set prediction variables (X)
iowa_prediction_variables = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X_iowa = home_data[iowa_prediction_variables]
```

```python
# Review data
# print description or statistics from X
print(X_iowa.describe())

# print the top few lines
print(X_iowa.head())
```

```python
# Use DecisionTreeRegressor and fit the model

iowa_model = DecisionTreeRegressor(random_state=1)

# Fit model
iowa_model.fit(X_iowa, y_iowa)
```

```python
# Make predictions for head of Iowa data
print('First in-sample predictions:', iowa_model.predict(X_iowa.head()))

print('Actual target values for those homes:', y_iowa.head().tolist())
```

```python
# Model Validation of Melbourne data, how well did our model predict the actual sale price

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

melbourne_housing_file_path = './melb_data.csv'
melb_data = pd.read_csv(melbourne_housing_file_path)
melb_data = melb_data.dropna(axis=0)
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 'YearBuilt', 'Lattitude', "Longtitude"]
X = melb_data[melbourne_features]
y = melb_data.Price

# split our data into training and test data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)

# Define our model, DecisionTreeRegressor
melbourne_model = DecisionTreeRegressor()

# Fit our model
melbourne_model.fit(train_X, train_y)

# Get predictions on our validation data
val_predictions = melbourne_model.predict(val_X)

print(mean_absolute_error(val_y, val_predictions))
```
